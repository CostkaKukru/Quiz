```python
import random

class Question:
    """
    Represents a quiz question, which includes a prompt and multiple choice answers.
    """
    def __init__(self, text, choices, answer, difficulty): # (dictionary questions = questions["text"], questions["choices"], questions["answer"], questions["difficulty"])
        self.text = text
        self.choices = choices
        self.answer = answer
        self.difficulty = difficulty

    def ask_question(self, i): # Adding time to answer each question?
        randomized_answer = random.sample(self.choices, len(self.choices))
        answer_dict = {}
        abcd = ["a", "b", "c", "d"]
        for j in range(len(self.choices)):  # Assigning abcd to the randomized_answer list
            answer_dict[abcd[j]] = randomized_answer[j]

        print(f"{i+1}. Question: {self.text}\na. {randomized_answer[0]}\nb. {randomized_answer[1]}\nc. {randomized_answer[2]}\nd. {randomized_answer[3]}")
        user_answer = input("Enter your answer (a, b, c, d): ").strip().lower()
        # Looping through the answer_dict to check if the user_answer is in the dictionary
        for key, value in answer_dict.items():
            if user_answer == key:
                user_answer = value
            else:
                pass

        return user_answer

class Quiz:
    """
    Manages a collection of Question objects and the process of taking the quiz.
    """
    def __init__(self, questions):
        self.questions = questions
        self.score = 0

    def introduction(self):
        print(f"Welcome to the Technical Trivia Quiz!\nAnswer with a, b, c or d.")

    def update_score(self, user_answer, correct_answer, points):
        if user_answer == correct_answer:
            self.score += points
            print("Correct! Great!")
            return True
        else:
            print("Sorry, incorrect.")
            return False

    def administer(self, number_of_questions):
        self.introduction()
        i = 0
        total = 0
        correct = 0  # Initialize the correct variable
        randomized_question = random.sample(self.questions, len(self.questions))
        for question in randomized_question:
            user_answer = Question.ask_question(question, i)
            if self.update_score(user_answer, question.answer, question.difficulty):
                correct += 1  # Increment the correct variable if the answer is correct
            total += question.difficulty
            i += 1
            if i == number_of_questions:
                break
            else:
                continue
        print(f"Your final score is: {self.score}/{total}")
        return self.score, total, correct  # Return the correct variable
```
